```python
# Function Definations

def fireWrite(Ref, Data, id):
    Ref.document(id).set(Data)


def fireRead(Ref, id):
    doc_ref = Ref.document(id)
    doc = doc_ref.get()
    return doc.to_dict()


def fireUpdate(Ref, Data, id):
    Ref.document(id).update(Data)
```

```python

#Algorithm that will determine Health Score of an individuals

def healthScore(patient):
    base = 100
    age = patient['Age']
# for the age below 0 to 18
    if(age <=18 and age>0):
        diseases = patient['Diseases']
        preExisitingDiseases = diseases['PreExisitingDiseases']
        if(preExisitingDiseases):
            preExisitingScale = diseases['PreExisitingScale']
            if(preExisitingScale>8):
                base = base - 5
            elif(preExisitingScale<=8 and preExisitingScale>6):
                base = base - 2
        
        familyDiseases = diseases['FamilyDiseases'] 
        if(familyDiseases):
            familyDiseasesScale = diseases['FamilyDiseasesScale']
            if(familyDiseasesScale>8):
                base = base-5
            elif(familyDiseasesScale<=8 and familyDiseasesScale>6):
                base = base-1
        bmi = patient['BodyMassIndex']
        if(bmi>24.9 or bmi < 18.5):
            base = base - 2
#for the age between 18 to 60
    if(age>18 and age<=60):
        diseases = patient['Diseases']
        preExisitingDiseases = diseases['PreExisitingDiseases']
        if(preExisitingDiseases):
            preExisitingScale = diseases['PreExisitingScale']
            if(preExisitingScale>8):
                base = base - 7
            elif(preExisitingScale<=8 and preExisitingScale>6):
                base = base - 3
        familyDiseases = diseases['FamilyDiseases']   
        if(familyDiseases):    
            familyDiseasesScale = diseases['FamilyDiseasesScale']
            if(familyDiseasesScale<=8 and familyDiseasesScale>6):
                base = base-3
            elif(familyDiseasesScale>8):
                base = base - 7
        bmi = patient['BodyMassIndex']
        if(bmi>24.9 or bmi < 18.5):
            base = base - 6
        habit = patient['Habits']
        smoker = habit['Smoking']
        if(smoker):
            smokerScale = habit['SmokingScale']
            if(smokerScale==3): #0,1,2,3
                base = base - 25
            if(smokerScale==2):
                base = base -15
            if(smokerScale == 1):
                base = base - 7
        alcoholic = habit['Alcoholic']
        if(alcoholic== True):
            alcoholicScale = habit['AlcoholicScale']
            if(alcoholicScale==0):
                base = base
            elif(alcoholicScale==3):
                base = base - 16
            elif(alcoholicScale==2):
                base = base - 10
            elif(alcoholicScale==1 ):
                base = base - 5
        tobacco = habit['Tobacco']
        if(tobacco== True):
            tobaccoScale = habit['TobaccoScale']
            if(tobaccoScale==0):
                base = base
            elif(tobaccoScale==3):
                base = base - 25
            elif(tobaccoScale==2):
                base = base - 15
            elif(tobaccoScale==1 ):
                base = base - 7
        

    if(age>60):
        diseases = patient['Diseases']
        preExisitingDiseases = diseases['PreExisitingDiseases']
        if(preExisitingDiseases):
            preExisitingScale = diseases['PreExisitingScale']
            if(preExisitingScale>8):
                base = base - 10
            elif(preExisitingScale<=8 and preExisitingScale>6):
                base = base - 5
        
        familyDiseases = diseases['FamilyDiseases']     
        if(familyDiseases):
            familyDiseasesScale = diseases['FamilyDiseasesScale']
            if(familyDiseasesScale>8):
                base = base - 10
            elif(familyDiseasesScale<=8 and familyDiseasesScale>6):
                base = base-6
        bmi = patient['BodyMassIndex']
        if(bmi>24.9 or bmi < 18.5):
            base = base - 10

        habit = patient['Habits']
        smoker = habit['Smoking']
        if(smoker):
            smokerScale = habit['SmokingScale']
            if(smokerScale==3): #0,1,2,3
                base = base - 35
            elif(smokerScale==2):
                base = base -20
            elif(smokerScale == 1):
                base = base - 10
        alcoholic = habit['Alcoholic']
        if(alcoholic):
            alcoholicScale = habit['AlcoholicScale']
            if(alcoholicScale==3):
                base = base - 20
            elif(alcoholicScale==2):
                base = base - 15
            if(alcoholicScale==1 ):
                base = base - 7
        tobacco = habit['Tobacco']
        if(tobacco== True):
            tobaccoScale = habit['TobaccoScale']
            if(tobaccoScale==0):
                base = base
            elif(tobaccoScale==3):
                base = base - 35
            elif(tobaccoScale==2):
                base = base - 20
            elif(tobaccoScale==1 ):
                base = base - 10
    print(base)
    return base
```
